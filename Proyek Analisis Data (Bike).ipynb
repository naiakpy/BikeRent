## Cleaning Data

### Delete Unused Data

```python
del hour_df
```

### Delete Unused Coloumn

```python
# List kolom yang akan dihapus
drop_cols = ["instant"]

# Iterasi melalui kolom-kolom dalam dataframe
for col in day_df.columns:
    # Periksa apakah nama kolom ada dalam list drop_cols
    if col in drop_cols:
        # Jika iya, hapus kolom dari dataframe
        day_df.drop(labels=col, axis=1, inplace=True)

# Menampilkan lima baris teratas dari dataframe setelah penghapusan kolom
day_df.head()
```

### Detailing Coloumn Name

```python
# Melakukan penggantian nama kolom dalam dataframe 'day_df'
day_df.rename(columns={
    "dteday": "dateday",
    "yr": "year",
    "mnth": "month",
    "weathersit": "weather_category",
    "hum": "humidity",
    "cnt": "count"
}, inplace=True)

# Menampilkan lima baris teratas dari dataframe setelah penggantian nama kolom
day_df.head()
```

### Breakdown of the counts in the columns for season, year, month, day, and weather

```python
# Convert numerical values to categorical labels for 'season' column
day_df["season"] = day_df["season"].map({
    1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"
})

# Convert numerical values to categorical labels for 'year' column
day_df["year"] = day_df["year"].map({
    0: "2011", 1: "2012"
})

# Convert numerical values to categorical labels for 'month' column
day_df["month"] = day_df["month"].map({
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun",
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
})

# Convert numerical values to categorical labels for 'weekday' column
day_df["weekday"] = day_df["weekday"].map({
    0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"
})

# Convert numerical values to categorical labels for 'weather_category' column
day_df["weather_category"] = day_df["weather_category"].map({
    1: "Clear/Partly Cloudy",
    2: "Misty/Cloudy",
    3: "Light Rain/Snow",
    4: "Heavy Rain/Snow"
})
```

### Change False Dataset

```python
# Convert 'dateday' column to datetime format
day_df["dateday"] = pd.to_datetime(day_df["dateday"])

# Convert other columns to categorical data type
day_df["season"] = day_df["season"].astype("category")
day_df["year"] = day_df["year"].astype("category")
day_df["month"] = day_df["month"].astype("category")
day_df["holiday"] = day_df["holiday"].astype("category")
day_df["weekday"] = day_df["weekday"].astype("category")
day_df["workingday"] = day_df["workingday"].astype("category")
day_df["weather_category"] = day_df["weather_category"].astype("category")

# Display the first few rows of the DataFrame
day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren penyewaan sepeda tahun 2011 - 2012?

```python
day_df["month"] = pd.Categorical(day_df["month"], categories=
    ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    ordered=True)

# Group data by year and month, aggregate total counts, and reset index
monthly_counts = day_df.groupby(by=["year", "month"]).agg({
    "count": "sum"
}).reset_index()

# Set the style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
# Create line plot
sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="Paired",
    marker="o"
)
plt.xlabel(None)
plt.ylabel(None)
plt.title("Bike Rental Trends in 2011 and 2012")
plt.legend(title="Year", loc="upper right")
plt.gca().xaxis.grid(False)
plt.show()
```

### Pertanyaan 2: Bagaimana perbandingan antara user yang register dan casual?

```python
user_counts = day_df.groupby(by="year").agg({
    "casual": "sum",
    "registered": "sum"
}).reset_index()

# Set the style
sns.set(style="whitegrid")

# Create a figure and axis
plt.figure(figsize=(10, 5))

# Plot the data using seaborn"s barplot
sns.barplot(data = user_counts, x = "year", y="casual", color="skyblue", label="Casual Users")
sns.barplot(data = user_counts, x = "year", y="registered", color="pink", label="Registered Users", bottom=user_counts["casual"])

# Add label and title
plt.xlabel(None)
plt.ylabel(None)
plt.title("Comparison of Bike Rental between Casual and Registered Users")

# Add a legend
plt.legend(title = "User Type", loc = "upper right")

plt.grid(False)
plt.show()
```

### Pertanyaan 3: Apakah ada perbedaan jumlah sewa sepeda pada weekdays dan weekend?

```python
weekday_counts = day_df.groupby(by="weekday").agg({
    "count": "sum",
}).reset_index()
weekday_counts_sorted = weekday_counts.sort_values(by="count", ascending=False)

sns.set(style="whitegrid")

fig_weekday, ax = plt.subplots(figsize=(24, 10))

sns.barplot(data = weekday_counts_sorted, x = "weekday", y = "count", color = "lightcoral")
plt.title("Total of Bike Rentals by Weekday", fontsize=21)
plt.xlabel(None)
plt.ylabel(None)

plt.grid(False)
plt.show()
```

### Pertanyaan 4: Bagaimana hubungan suhu dan cuaca dengan jumlah jumlah penyewaan sepeda?

```python
plt.figure(figsize=(12, 6))

# Scatter plot for "temp" vs "count"
plt.subplot(2, 2, 1)
sns.scatterplot(
    x="temp",
    y="count",
    data=day_df,
    alpha=0.5
)
plt.title("Temperature vs Count")

# Scatter plot for "atemp" vs "count"
plt.subplot(2, 2, 2)
sns.scatterplot(
    x="atemp",
    y="count",
    data=day_df,
    alpha=0.5
)
plt.title("Temperature (Feels Like) vs Count")

# Scatter plot for "humidity" vs "count"
plt.subplot(2, 2, 3)
sns.scatterplot(
    x="humidity",
    y="count",
    data=day_df,
    alpha=0.5
)
plt.title("Humidity vs Count")

# Scatter plot for "windspeed" vs "count"
plt.subplot(2, 2, 4)
sns.scatterplot(
    x="windspeed",
    y="count",
    data=day_df,
    alpha=0.5
)
plt.title("Windspeed vs Total Rental Bikes")

plt.tight_layout()
plt.show()
```
